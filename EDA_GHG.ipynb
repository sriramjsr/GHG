```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('final_datas.xlsx')
```

```python
df
```

```python
df.columns
```

```python
df.drop(columns=['Natural gas(CM)'],inplace=True)
```

```python
df.isnull().sum()/len(df) * 100
```

```python
df.isnull().sum()/len(df) *100
```

```python
k = df[df.duplicated]
```

```python
df.drop(k.index,inplace=True)
```

```python
df
```

```python
da=['Electricity(KWH)','Natural Gas(CM)','Propane(L)','Coal Quantity(MT)','Wood(MT)','District Heating(GJ)','District Cooling(GJ)','Fuel_Oil(L)']
for i in da:
    df[i].replace(np.nan,0,inplace=True)
```

```python
df.isnull().sum()/len(df) * 100
```

```python
df['Average_Hours'] = df['Average_Hours'].fillna(df['Average_Hours'].median())
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.to_excel('ghg_clean_data.xlsx',index=False)
```
